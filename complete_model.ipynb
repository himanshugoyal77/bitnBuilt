```python
df= df[["age", "Medu", "Fedu", "Mjob", "Fjob", "studytime", "failures", "higher", "address", "G3"]]
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns= ['Mjob', 'Fjob', 'higher', 'address'], drop_first=True)
```

```python
df.head()
```

```python
X = df.loc[:, df.columns!="G3"]
y = df["G3"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
sns.displot(df["G3"]);
```

# **Linear Regression**

```python

from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train.values, y_train)
```

```python
prediction = model.predict(X_test.values)
```

```python
np_y_test = np.array(y_test)
```

```python
np_y_test
```

```python
prediction[30]
```

```python
np_y_test[30]
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(y_test, prediction)
```

```python
for ind in range(0, 20):
  np_y_test = np.array(y_test)
  actual = np_y_test[ind]
  pred = round(prediction[ind], 2)
  diff = round(actual - pred, 2)
  prec = round(diff/actual * 100, 2)
  print("Actual : " + str(actual))
  print("Prediction : " +  str(pred))
  print("Error : " + str(diff))
  print("Error % : " + str(prec))
```

```python
print("Root mean square error : ", np.sqrt(mean_squared_error(y_test, prediction)))
```
